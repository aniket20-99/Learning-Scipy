```python
# spare matrix methods;viewing stored data with the data property
import numpy as np
from scipy.sparse import csr_matrix
y = np.array([[0,0,0],[0,0,2],[1,0,4]])
print(csr_matrix(y).data)
```

```python
#counting non zero with the count_nonzero(method)
import numpy as np
from scipy.sparse import csr_matrix
y = np.array([[0,0,0],[0,0,2],[1,0,4]])
print(csr_matrix(y).count_nonzero())
```

```python
# removing zero entries from the matrix with the eliminate_zeros(method)
import numpy as np
from scipy.sparse import csr_matrix
y = np.array([[0,0,0],[0,0,2],[1,0,4]])
x = csr_matrix(y)
x.eliminate_zeros()
print(x)
```

```python
#sum_duplicates 
import numpy as np
from scipy.sparse import csr_matrix
y = np.array([[0,0,0],[0,0,2],[1,0,4]])
x = csr_matrix(y)
x.sum_duplicates()
print(x)
```

```python
#convering from csr to csc with the tocsc(method)
import numpy as np
from scipy.sparse import csr_matrix
y = np.array([[0,0,0],[0,0,2],[1,0,4]])
y1 = csr_matrix(y).tocsc()
print(y1)
```

```python
# Spatial data refers to data that is represented in geomatrics space its deals with spatial data problem on many tasks, scipy provides us module scipy.spatial which has function for working with spatial data
# Triangulisation -- polygon is to devide the polygon * multi triangle which we can compute area of polygon.
```

```python
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

points = np.array([
    [2,4],
    [3,5],
    [5,8],
    [2,6],
    [4,5]
])

simplices = Delaunay(points).simplices
plt.triplot(points[0:,0],points[0:,1],simplices)
plt.scatter(points[:,0],points[:,1],color = 'red')
plt.show()
```

```python
# Graph --> adjascent matrix
import numpy as np
from scipy.sparse.csgraph import connected_components
from scipy.sparse import csr_matrix
arr = np.array([
    [0,1,2],
    [1,0,0],
    [2,0,0]
])

new_arr = csr_matrix(arr)
print(connected_components(new_arr))
```

```python
#Floydwarshall(method to find shortest path between all pairs of elements)
import numpy as np
from scipy.sparse.csgraph import floyd_warshall
from scipy.sparse import csr_matrix
arr = np.array([
    [0,1,2],
    [1,0,0],
    [2,0,0]
])

new_arr = csr_matrix(arr)
print(floyd_warshall(new_arr,return_predecessors = True))
```

```python
#1.Write a NumPy program that creates a NumPy array of random numbers and uses SciPy to compute the statistical properties (mean, median, variance) of the array.
import numpy as np
from scipy import stats

data = np.random.rand(100)
mean = stats.tmean(data)
median = np.median(data)
variance = stats.tvar(data)
print('mean:',mean)
print('median:',median)
print('variance:', variance)
```

```python
#2.Write a NumPy program to generate a 2D array and performs a discrete Fourier transform using SciPy's fftpack module.

import numpy as np
from scipy import linalg

matrix = np.array([[4,2],[3,1]])
# find the determinant of the matrix using scipy
determinant = linalg.det(matrix)
# find the inverse of the matrix using matrix
inverse = linalg.inv(matrix)

# find the eignvalue of the matrix using scipy
eignvalues,eignvectors = linalg.eig(matrix)

print('matrix:',matrix)
print('determinant:',determinant)
print('inverse:',inverse)
print('eignvalues:',eignvalues)
print('eignvectors:',eignvectors)
```

```python
#3.Write a NumPy program to create a time series dataset and apply SciPy's signal processing functions to filter the data.
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
np.random.seed(0)
time = np.linspace(0,1,500,endpoint = False)
data = np.sin(2*np.pi*7*time) + np.random.randn(500)*0.5
b,a = signal.butter(4,0.1)
filter_data = signal.filtfilt(b,a,data)
plt.plot(time,data,label = 'original data')
plt.plot(time,filter_data,label = 'filter_data',linewidth = 2)
plt.legend()
plt.show()
```
